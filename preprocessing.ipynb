## Cleaning

```python
# Columns
house_data.columns
```

```python
# Remove duplicates
house_data = house_data.drop_duplicates()
```

### Typecasting

```python
house_data['date'] = pd.to_datetime(house_data['date'].str.slice(0,8), format='%Y%m%d')
# house_data['date'] = house_data['date'].dt.strftime('%d%m%Y')
```

```python
house_data['date']
```

```python
# Type cast bathroom to int
house_data['bathrooms'] = np.floor(house_data['bathrooms'])
house_data['floors'] = np.floor(house_data['floors'])
```

```python
# Look at the date distribution
sns.kdeplot(house_data['date'])
```

```python
# Type cast 'bathrooms' and 'floors' to int
house_data['bathrooms'] = house_data['bathrooms'].astype(int)
house_data['floors'] = house_data['floors'].astype(int)
```

```python
# Convert the areas to metrics system
house_data['sqft_living'] = house_data['sqft_living']/10.764

house_data['sqft_lot'] = house_data['sqft_lot']/10.764

house_data['sqft_above'] = house_data['sqft_above']/10.764

house_data['sqft_basement'] = house_data['sqft_basement']/10.764

house_data['sqft_lot15'] = house_data['sqft_lot15']/10.764

house_data['sqft_living15'] = house_data['sqft_living15']/10.764
```

```python
house_data = house_data.rename(columns={
    'sqft_living': 'living',
    'sqft_lot': 'lot',
    'sqft_above': 'above',
    'sqft_basement': 'basement',
    'sqft_living15': 'living15',
    'sqft_lot15': 'lot15'
})
```
