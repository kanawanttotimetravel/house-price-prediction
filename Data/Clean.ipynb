CLEAN

```python
#Lọc các cột không cần thiết
pd1 = pd1.drop(['waterfront','sqft_basement','zipcode'], axis = 1)
```

```python
pd1.columns
```

```python
#Lọc duplicate
pd1 = pd1.drop_duplicates()
```

```python
#Chuẩn hóa dữ liệu ngày thành dạng dd/mm/yyyy
pd1['date'] = pd.to_datetime(pd1['date'].str.slice(0,8), format='%Y%m%d')
pd1['date'] = pd1['date'].dt.strftime('%d%m%Y')
```

```python
pd1['date']
```

```python
#Làm tròn dữ liệu bathroom
pd1['bathrooms'] = np.floor(pd1['bathrooms'])
```

```python
pd1['bathrooms']
```

```python
# Ép kiểu dữ liệu của cột 'bathrooms' và 'floors' về int
pd1['bathrooms'] = pd1['bathrooms'].astype(int)
pd1['floors'] = pd1['floors'].astype(int)
```

```python
pd1['bathrooms']
```

```python
pd1['floors']
```

```python
#Lọc outlier

#Chuẩn hóa dữ liệu diện tích khu nhà sang m^2
pd1['sqft_living'] = pd1['sqft_living']/10.764

#Chuẩn hóa dữ liệu diện tích đất sang m^2
pd1['sqft_lot'] = pd1['sqft_lot']/10.764

#Chuẩn hóa dữ liệu diện tích trong nhà sang m^2
pd1['sqft_above'] = pd1['sqft_above']/10.764

#Chuẩn hóa dữ liệu diện tích đất hàng xóm sang m^2
pd1['sqft_lot15'] = pd1['sqft_lot15']/10.764

#Chuẩn hóa dữ liệu diện tích khu nhà hàng xóm sang m^2
pd1['sqft_living15'] = pd1['sqft_living15']/10.764
```

```python
from sklearn.model_selection import train_test_split

# Chia df1 thành train và temp với tỉ lệ 8:2
train, temp = train_test_split(pd1, test_size=0.2, random_state=42)

# Chia temp thành test và valid với tỉ lệ 1:1
test, valid = train_test_split(temp, test_size=0.5, random_state=42)

# Xuất DataFrame train ra file CSV
train.to_csv('train.csv', index=False)

# Xuất DataFrame valid ra file CSV
valid.to_csv('valid.csv', index=False)

# Xuất DataFrame test ra file CSV
test.to_csv('test.csv', index=False)
```
